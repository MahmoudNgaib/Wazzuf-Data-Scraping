```python
#importing libararies
import pandas as pd
from bs4 import BeautifulSoup
import requests
import urllib.request, urllib.error, urllib.parse
import math
```

* Define a function to read search result page

```python
def get_url (query):
    parms = {'a':'spbg', 'q':query}
    url = str('https://wuzzuf.net/search/jobs/?{}'.format(urllib.parse.urlencode(parms)))
    return url
```

```python
def get_soup(url):
    html = requests.get(url)
    return BeautifulSoup(html.content, 'html.parser')
```

* getting total number of jobs 

```python
def get_total_jobs(soup):
    page_summary_element = soup.find(attrs={"class": "css-7o92qm"})
    if not page_summary_element:
        return 0
    page_summary = page_summary_element.find("li", attrs={"class": "css-8neukt"})
    if not page_summary:
        return 0
    return int(page_summary.text.partition('of')[-1].strip())
```

* Get how many Jobs in a page

```python
def get_jobs_per_page(soup):
    mainpage_jobs_divs = soup.find(attrs={"class": "css-9i2afk"}).find_all(attrs={"class": "css-1gatmva e1v1l3u10"})
    return len(mainpage_jobs_divs)
```

* now Create a list that reads all info in a page

```python
def parse_job_div(one_div):
    job_data = {}
    
    job_heading = one_div.find('h2', attrs={"class": "css-m604qf"})
    job_data['Job Heading'] = job_heading.text if job_heading else None
    
    job_status = one_div.find(attrs={"class": ["css-4c4ojb", "css-do6t5g"]})
    job_data['Job Status'] = job_status.text if job_status else None
    
    company_name = one_div.find(attrs={"class": "css-d7j1kk"}).text.split('-')[0].strip()
    job_data['Company Name'] = company_name if company_name else None
    
    country = one_div.find(attrs={"class": "css-5wys0k"}).text.split(',')[-1].strip()
    adress = one_div.find(attrs={"class": "css-5wys0k"}).text.partition(country)[0]
    adress_cleared = adress[:len(adress)-2]
    job_data['Job Adress'] = adress_cleared if adress_cleared else None
    
    job_description_divs = one_div.find(attrs={"class": "css-y4udm8"})
    job_data['Job Description'] = job_description_divs.text if job_description_divs else None
    
    job_link = one_div.find(attrs={"class": "css-m604qf"}).find('a')
    job_data['Job Link'] = job_link['href'] if job_link else None
    
    return job_data
```

* Now let's parse all pages

```python
def read_webPage(query):
    url = get_url(query)
    soup = get_soup(url)
    tot_nom_jobs = get_total_jobs(soup)
    num_jobs_on_page = get_jobs_per_page(soup)
    no_pages = min(math.ceil(tot_nom_jobs / num_jobs_on_page), 7)  # Limit to 7 pages....hhh if it is more than 7 pages it makes may wanna stop programming hhh    
    job_data_list = []
    
    for page in range(no_pages):
        url_page = f"{url}&{urllib.parse.urlencode({'start': page})}"
        soup_page = get_soup(url_page)
        
        jobs_divs_page = soup_page.find(attrs={"class": "css-9i2afk"})
        if not jobs_divs_page:
            continue
        
        jobs_div_page = jobs_divs_page.find_all(attrs={"class": "css-1gatmva e1v1l3u10"})
        for one_div in jobs_div_page:
            job_data_list.append(parse_job_div(one_div))
    
    df = pd.DataFrame(job_data_list)
    return df
```

```python
query = "data scientist"
read_webPage(query)
```
